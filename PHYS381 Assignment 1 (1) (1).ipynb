```python
import numpy as np
import scipy.stats as ss
import matplotlib as mp
import matplotlib.pyplot as plt

A = np.loadtxt("SampleA.dat")
B = np.loadtxt("SampleB.dat")

mean_A = np.mean(A)
mean_B = np.mean(B)


var_A = np.var(A)
var_B = np.var(B)

t_test = ss.ttest_ind(A, B, equal_var=False)

t_value = t_test[0]
print(t_value)


p_value = t_test[1]
print(p_value)
#strong evidence for the null hypothesis, not statistically significant. So the difference between the means is not statistically significant?
#i.e. our samples are quite similar.

hist_A = plt.hist(A, bins=20, alpha=0.5, color='black')
hist_B = plt.hist(B, bins=20, alpha=0.5, color='yellow')
#plotting the sample data with two overlapping histograms with transparent bins

plt.axvline(mean_A)
plt.axvline(mean_B)
#through both methods we are able to show that the difference between the means is not statistically significant

```
